<h3>Data Cleaning and Preprocessing</h3>

```python
# Handling missing value
zomato["rate"] = zomato["rate"].replace(["NEW", "-"], np.nan)
zomato["rate"] = zomato["rate"].str.replace("/5"," ").astype(float)
zomato["rate"].dtype
```

```python
zomato.head(2)
```

```python
# Check number of missing values in the ‘rate’ column
zomato["rate"].isnull().sum()
```

```python
# Remove rows where ‘rate’ is missing
zomato.dropna(subset = ["rate"], inplace = True)
zomato["rate"].isnull().sum()
```

```python
# Fill missing locations using the ‘listed_in(city)’ column
zomato["location"] = zomato["location"].fillna(zomato["listed_in(city)"])
```

```python
# Replace missing restaurant types with ‘Unknown'
zomato["rest_type"] = zomato["rest_type"].fillna("Unknown")
zomato["rest_type"].isnull().sum()
```

```python
# Fill missing values in ‘dish_liked’ and ‘cuisines’ with ‘Not available'
zomato[["dish_liked", "cuisines"]] = zomato[["dish_liked", "cuisines"]].fillna("Not available")
```

```python
# Clean cost column by removing commas and converting to float
zomato["approx_cost(for two people)"] = zomato["approx_cost(for two people)"].str.replace(",","").astype(float)
```

```python
zomato["approx_cost(for two people)"]
```

```python
# Fill missing cost values using group median by rest_type and location
zomato["approx_cost(for two people)"] = zomato["approx_cost(for two people)"].fillna(zomato.groupby(["rest_type", "location"])["approx_cost(for two people)"].transform("median"))
zomato.loc[49373] # View updated row 49373
```

```python
# Check missing values in 'approx_cost(for two people)' 
zomato["approx_cost(for two people)"].isnull().sum()
```

```python
#dropping the columns
zomato.drop(["menu_item", "reviews_list", "phone"],axis=1)
zomato.head(4)
```

<h4>4.Cost Vs Rating</h4>

```python
# 4. Cost Vs Rating
max = zomato["approx_cost(for two people)"].max()
min = zomato["approx_cost(for two people)"].min()
avg = zomato["approx_cost(for two people)"].mean()
print("Max: ", max)
print("Min: ", min)
print("Average: ", avg)
```

```python
bins = [0, 300, 600, 1000, 3000, 6000]
labels = ["Budget", "Affordable", "Mid-range", "Premium", "Luxury"]
zomato["cost_bins"] = pd.cut(zomato["approx_cost(for two people)"], bins = bins, labels = labels, right = False)
zomato.head(2)
```

```python
#Visualization -- box plot
plt.figure(figsize = (12, 6), facecolor = "white")
zomato.boxplot(column = "rate", by = "cost_bins")
plt.xlabel("Cost Category", labelpad = 15)
plt.ylabel("Rating", labelpad = 15)
plt.title("Rating Distribution by Cost Range", pad = 15)
plt.style.use("default")
plt.suptitle("")
plt.tight_layout()
plt.savefig("Rating Distribution", dpi = 100)
plt.show()
```

<p>Premium and Luxury restaurants consistently receive the highest ratings, while Budget and Affordable categories show more variability with lower median ratings</p>

<h4>5.Restaurants with high votes but low ratings</h4>

```python
# 5.Restaurants with higher votes but low rates
high_votes = zomato["votes"].quantile(0.75)
low_rates = zomato["rate"].mean()
high_votes_low_rates = zomato[(zomato["votes"] > high_votes) & (zomato["rate"] < low_rates)]
high_votes_low_rates.head(4)
```

```python
#Visualization -- Scatter plot
plt.figure(figsize = (10, 6))

# All restaurants (faded)
plt.scatter(zomato["votes"], zomato["rate"], alpha = 0.2, label = "All Restaurants")

# Suspicious ones
plt.scatter(
    high_votes_low_rates["votes"],
    high_votes_low_rates["rate"],
    color = "red",
    label = "High Votes + Low Rating"
)

plt.axhline(low_rates, color = "gray", linestyle = "--")  # rating threshold
plt.axvline(high_votes, color = "gray", linestyle = "--") # votes threshold

plt.xlabel("Votes", labelpad = 15)
plt.ylabel("Rating", labelpad = 15)
plt.title("Restaurants With High Votes but Low Ratings", pad = 15)
plt.legend()
plt.style.use("dark_background")
plt.tight_layout()
plt.savefig("Restaurants", dpi = 100)
plt.show()
```

<p>Some restaurants attract many votes but still maintain low ratings, revealing a mismatch between popularity and quality</p>

<h4>6.Top Restaurants with High Ratings and Low Costs on the Map</h4>

```python
# Locating Top Restaurants with higher ratings and low costs on the map
df_unique = zomato.groupby(['name','address','listed_in(city)'], as_index=False).agg({
    'rate': 'mean',
    'votes': 'max',
    'approx_cost(for two people)': 'mean'
})
areas = df_unique['listed_in(city)'].dropna().unique()
len(areas)
##
from geopy.geocoders import Nominatim
import pandas as pd
import time

geolocator = Nominatim(user_agent="zomato_area_mapping")

area_latlng = {}

for area in areas:
    try:
        location = geolocator.geocode(f"{area}, Bangalore, India", timeout=10)
        time.sleep(1)   # VERY IMPORTANT (API rate limit)
        
        if location:
            area_latlng[area] = (location.latitude, location.longitude)
        else:
            area_latlng[area] = (None, None)
    except:
        area_latlng[area] = (None, None)
area_coords_df = pd.DataFrame.from_dict(
    area_latlng,
    orient='index',
    columns=['lat','lng']
).reset_index().rename(columns={'index':'listed_in(city)'})
df_unique = df_unique.merge(area_coords_df, on='listed_in(city)', how='left')
##
low_cost_threshold = 300
high_rating_threshold = df_unique['rate'].mean()

top10 = (
    df_unique[
        (df_unique['approx_cost(for two people)'] <= low_cost_threshold) &
        (df_unique['rate'] >= high_rating_threshold)
    ]
    .sort_values(by='rate', ascending=False)
    .head(10)
)
##
import folium

bangalore_map = folium.Map(location=[12.97, 77.59], zoom_start=12)

for _, row in top10.dropna(subset=['lat','lng']).iterrows():
    folium.Marker(
        location=[row['lat'], row['lng']],
        popup=f"""
        <b>{row['name']}</b><br>
        Area: {row['listed_in(city)']}<br>
        Rating: {row['rate']}<br>
        Cost: ₹{row['approx_cost(for two people)']}
        """,
        icon=folium.Icon(color='green')
    ).add_to(bangalore_map)

bangalore_map.save("map.html")
```
